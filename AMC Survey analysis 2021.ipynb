```python
# grouping by mail sum total spent
#one id per mail, missing 20 cases but seems that unique cases
customer_AMC.groupby('Email')['Customer ID'].count()
```

```python
#merging with survey data
survey_customer_AMC = pd.merge(customer_AMC, survey_AMC, how="outer", left_on='Email', right_on='Email')
```

```python
survey_customer_AMC.to_csv(r'annualsurvey2021_customerdatametorikAMC.csv')
```

```python
len(survey_customer_AMC)
```

```python
#Is any of the questions of the survey significant to total spent? random forest regressor
#filter the ones that answered the survey
survey_answered = survey_customer_AMC['Age'] != 'NaN'
survey_answered_df = survey_customer_AMC[survey_customer_AMC.Age.notnull()] #filtering by not null
```

```python
survey_answered_df.isnull().sum()
```

```python
#deliting duplicates
survey_answered_df['Email'].nunique()
```

```python
len(survey_answered_df['Email'])
```

```python
#deliting duplicates keeping unique mails first case
survey_answered_df.drop_duplicates(subset ="Email",
                     keep = "first", inplace = True)
```

```python
grouped_multiple = survey_answered_df.groupby(['Billing Address Country', 'What is your primarily spoken language?' ]).agg({'Total Spent': ['mean']})
```

```python
grouped_multiple = grouped_multiple.reset_index()
```

```python
grouped_multiple.to_csv(r'totalspent_bycountry_bylenguage_AMC.csv')
```

```python
grouped_profession = survey_answered_df.groupby(['What is your Occupation?' ]).agg({'Total Spent': ['mean']})
```

```python
grouped_profession = grouped_profession.reset_index()
```

```python
grouped_profession.to_csv(r'avgspent_byprofession_AMC.csv')
```

```python
grouped_proff_reinbursement = survey_answered_df.groupby(['What is your Occupation?', 'Does your employer compensate you for the cost of your certification course?' ]).agg({'Total Spent': ['mean']})
```

```python
grouped_proff_reinbursement = grouped_proff_reinbursement.reset_index()
```

```python
grouped_proff_reinbursement.to_csv(r'grouped_proff_reinbursement_AMC.csv')
```
